## Analysis 1️⃣: Number of books per section

To begin our analysis, we start by examining the distribution of books across different library sections. This gives us a clear overview of how the collection is organized and which topics or areas are most represented.

We will use a **pie chart** to visualize the number of books in each section, which allows us to quickly identify:

- The most and least populated sections
- How balanced the library's content is across different domains
- Potential areas for expanding the collection

This high-level insight helps set the stage for deeper performance-based analysis in the following sections.


```python
section_counts = df["section"].value_counts().reset_index()
section_counts.columns = ["section", "count"]

fig = px.pie(data_frame=section_counts, names="section", values="count", title="Number of Books per Section")

fig.show()
```

The **Fiction** section overwhelmingly dominates the library's collection, accounting for **over half ** of all books. This suggests a strong emphasis on fictional literature within the library’s holdings.

Other notable sections include **Children**, **Humanities**, and **Non-Fiction**, which are also well-represented. On the other hand, sections such as **Science**, **History**, and **Reference** hold a relatively small portion of the total collection, indicating potential areas for future growth or reevaluation.


## Analysis 2️⃣: Average Borrowings per Section

In this section, we explore how frequently books are borrowed in each library section by calculating the **average number of borrowings per book**.

By visualizing this data with a **bar chart**, we can quickly identify which sections are the most actively read by users, and which may receive less attention.

This kind of insight is valuable for:
- Understanding reader preferences across different subject areas
- Informing decisions about which sections may need more promotion or expansion
- Detecting underutilized parts of the library collection

Each bar represents the average borrowings for all books in a particular section, offering a clear picture of section-level performance.


```python
avg_borrowed_per_section = df.groupby(
    "section"
)["times_borrowed"].mean().reset_index()

avg_borrowed_per_section.columns = ["section", "avg_times_borrowed"]

df_sorted = avg_borrowed_per_section.sort_values("avg_times_borrowed", ascending=False)

fig = px.bar(
    df_sorted,
    x="section",
    y="avg_times_borrowed",
    title="Average Number of Borrowings per Section"
)

fig.update_layout(
    xaxis_title="Section",
    yaxis_title="Average Borrowings",
    xaxis_tickangle=45
)

fig.show()
```

The chart reveals clear differences in reader engagement across library sections. 

The **Young Adult** section stands out with the highest average number of borrowings per book, suggesting that this category is extremely popular among library users. **Non-Fiction** also performs well, with a significantly higher borrowing average than most other sections.

On the other hand, sections such as **Humanities**, **History**, and **Reference** show relatively low borrowing activity, which may indicate lower reader interest or accessibility issues.

These insights can help inform decisions about resource allocation, promotion strategies, or content curation within specific sections of the library.


## Analysis 3️⃣: Distribution of Book Ratings per Section

In this section, we examine how book ratings are distributed across different library sections. Understanding rating patterns can provide valuable insight into:

- The perceived quality of books within each section
- Reader satisfaction by subject area
- Potential outliers or inconsistencies in content

To visualize these patterns, we will use a **histogram** combined with a **box plot**:

- The histogram shows the **overall spread** and frequency of ratings per section.
- The box plot reveals **statistical characteristics** such as median, quartiles, and outliers for each group.

This dual approach allows us to both observe general trends and detect subtle differences in how readers evaluate books from different sections.


```python
fig = px.histogram(
    df,
    x="rating",
    color="section",
    marginal="box",
    hover_data=["title", "author", "rating", "ratings_count"],
    title="Distribution of Book Ratings"
)

fig.update_layout(
    xaxis_title="Rating",
    yaxis_title="Books (No.)",
    xaxis_range=[0, 5],
    bargap=0.1
)

fig.show()
```

The distribution of book ratings shows that most books are rated highly, with the largest concentration of ratings between **4.0 and 5.0**. This suggests generally positive feedback across the library collection.

- The **Fiction** section clearly has the largest number of books, which makes it the dominant part of the histogram. Most Fiction ratings are concentrated around the higher end of the scale, especially between **4.3 and 4.8**.
- Several sections, including **History**, **Children**, **Non-Fiction**, **Science**, **Humanities**, and **Young Adult**, also show ratings concentrated in a similar high-rating range.
- The histogram shows a small number of low-rated books, mostly below **3.5**, but these are relatively rare compared to the main concentration of highly rated books.
- The **box plots** indicate that most sections have relatively narrow interquartile ranges, meaning that ratings are fairly stable within each section.
- Some sections show individual outliers, especially on the lower end of the rating scale, which may indicate specific books that received noticeably weaker feedback.

Overall, the visualization suggests that the library collection is generally well-rated, with most sections showing strong and consistent reader satisfaction. The main difference between sections is not only in their average rating, but also in the number of books available in each section, with **Fiction** standing out as the largest category.

## Analysis 4️⃣: Relationship Between Publication Year and Borrowings by Section

In this analysis, we examine whether there is a relationship between a book’s year of publication (`year_published`) and the number of times it has been borrowed (`times_borrowed`) within different library sections.

By analyzing this relationship section by section, we aim to:

- Determine whether newer or older books tend to be borrowed more often
- Compare borrowing patterns across different parts of the library collection
- Identify sections where publication year may have a stronger influence on user interest
- Detect whether some sections show little or no relationship between book age and borrowing activity

This helps us understand whether borrowing behavior is connected to the age of books, and whether that connection differs depending on the section.


```python
fig = px.scatter(df,
                 x="year_published",
                 y="times_borrowed",
                 color="section",
                 trendline="lowess",
                 hover_name="title")

fig.update_layout(
    xaxis_title="Year published",
    yaxis_title="Times borrowed",
    yaxis_range=[0, 2500]
)

fig.show()
```

By analyzing each section individually, we observe distinct borrowing trends based on publication year and popularity:

#### 🔵 Fiction
- Fiction has the **highest density and peak borrowing values**, especially after the year **1980**.
- A smaller group of titles reaches very high borrowing values, including more than 500 borrowings.

#### 🟢 Children
- Borrowings in the Children section increase steadily after **1950**, with several popular titles borrowed **200+ times**.
- This suggests consistent interest and long-term engagement with children’s literature.

#### 🔴 History
- Borrowing activity is **low and flat** over time, even for more recent books.
- Indicates potential for improvement through **modernization or better promotion**.

#### 🔵 Humanities
- Exhibits **low borrowing counts across all periods** with no visible growth.
- This may indicate lower borrowing engagement and suggests that the section could benefit from further review.

#### 🟣 Non-Fiction
- Non-Fiction shows **moderate borrowing levels**, with a clear rise after **1990**.
- Recent books are performing better, signaling **growing user interest in factual or educational content**.

#### 🟢 Reference
- Rarely borrowed, even for newer books.
- May reflect that these materials are **used on-site rather than checked out**, or are **less appealing in a modern digital context**.

#### 🟠 Science
- Borrowing is limited but slightly increases in the **2000s**.
- A few standout books may be skewing the trend; consider highlighting high-value titles.

#### 🩷 Young Adult
- Consistent upward trend since **1980**, with several titles showing **very high borrowings**.
- Confirms that this section is **growing in popularity**, especially among newer generations.
