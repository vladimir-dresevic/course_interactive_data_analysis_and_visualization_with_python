# Interactive Data Analysis in Jupyter
This notebook demonstrates basic features of Jupyter notebooks for simple data-related tasks.

## 1. Sum of Two Numbers
Define two variables and display their sum.

```
a = 12
b = 8
sum_result = a + b
sum_result
```

## 2. Display a Message
Use the `print()` function to display a formatted message.

```
print(f"The sum of {a} and {b} is {sum_result}.")
```

## 3. List and Loop
Display the square of each number in a list using a `for` loop.

```
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(f"{number}^2 = {number ** 2}")
```

## 4. Average Value
Calculate and print the average value of the list.

```
average = sum(numbers) / len(numbers)
print(f"The average value is: {average}")
```

In this notebook, we practiced basic operations: addition, formatted printing, working with lists and loops, and computing averages.
